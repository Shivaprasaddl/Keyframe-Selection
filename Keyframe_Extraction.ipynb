## Function for Frame Extraction

```python
import cv2

def Frame_extraction(vpt):
    video = cv2.VideoCapture(vpt)
    fms=[]
    while True:
        ret, frame = video.read()
        if not ret:
            print('Finished.. Total: ',len(fms))
            break
        else:
            fms.append(frame)
    video.release()
    return fms
```

## Keyframe Selection Process

```python
from Duplicate_Reduction import Duplicates_Removal
import os

pt = './Dataset/' # path of the directory consist of all videos
sp = './Keyframe/' # path of directory to save keyframes

for i in os.listdir(pt):
    nm = i.split('.')[0] # taking name of video
    os.makedirs(sp+nm,exist_ok=True) # creating folder to store keyframes
    fms=Duplicates_Removal(Frame_extraction(pt+i),0.95).sel  # keyframe selection with removing 95% similar frames
    [cv2.imwrite(sp+nm'/{}_{}.jpg'.format(nm,c),j) for c,j in enumerate(fms)] # keyframe saving to corresponding folder
    print(i,' video keyframes extracted successfully...')   
```
